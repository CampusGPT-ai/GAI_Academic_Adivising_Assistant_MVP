```python
# import data from csv file
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/marynelson/GAI_Academic_Advising_Assistant_MVP/exported_messages.csv')
df = df[df['message_role'] == 'user']
df.head()
                      
```

```python
from conversation.run_graph import GraphFinder
from pathlib import Path
import json, csv
from data.models import UserSession

session_path = Path('/Users/marynelson/GAI_Academic_Advising_Assistant_MVP/app/data/mock_user_session.json')
with session_path.open(mode='r') as file:
    mock_user_session : UserSession = UserSession(**json.load(file))

results = []

def calc_similarity(question):
    graph_finder = GraphFinder(mock_user_session, question)
    topic_scores = graph_finder.get_topic_list_from_question()
    for topic_score in topic_scores:
        results.append([question, topic_score['name'], topic_score['score']])
    return results
```

```python
df['topic_match'] = df.apply(lambda x: calc_similarity(x['message_content']), axis=1)
df.head()
```
